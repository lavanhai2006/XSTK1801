```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 1. Dữ liệu từ bảng Excel của bạn
# X (Thu nhập), Y (Chi tiêu)
data = {
    "x": [8, 10, 12, 12, 14, 16, 16, 18, 20, 20], # Thu nhập
    "y": [8, 11, 11, 10, 14, 15, 11, 15, 14, 16]  # Chi tiêu
}

# 2. Chuyển dữ liệu thành DataFrame
df = pd.DataFrame(data)

# 3. Tách biến độc lập (X) và phụ thuộc (y)
X = df[["x"]].values # Định dạng 2D cho sklearn
y = df["y"].values

# 4. Xây dựng mô hình hồi quy tuyến tính
model = LinearRegression()
model.fit(X, y)

# 5. Dự đoán giá trị
y_pred = model.predict(X)

# 6. Tính toán các hệ số hồi quy (Khớp với kết quả Excel của bạn)
beta_0 = model.intercept_  # Hệ số chặn (Trong Excel là 4.548)
beta_1 = model.coef_[0]    # Hệ số dốc (Trong Excel là 0.5446)

# 7. Đánh giá mô hình
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)

# In kết quả
print(f"Phương trình hồi quy: y = {beta_0:.4f} + {beta_1:.4f}x")
print(f"Hệ số xác định (R^2): {r2:.4f}")
print(f"Sai số bình phương trung bình (MSE): {mse:.4f}")

# 8. Vẽ đồ thị
plt.scatter(X, y, color='blue', label='Dữ liệu thực tế')
plt.plot(X, y_pred, color='red', label='Đường hồi quy')

# --- THÊM DÒNG KẺ Ở ĐÂY ---
plt.grid(True, linestyle='--', alpha=0.7)
# --------------------------

plt.xlabel('Thu nhập (X)')
plt.ylabel('Chi tiêu (Y)')
plt.title('Hồi quy tuyến tính: Thu nhập và Chi tiêu')
plt.legend()
plt.show()
```
